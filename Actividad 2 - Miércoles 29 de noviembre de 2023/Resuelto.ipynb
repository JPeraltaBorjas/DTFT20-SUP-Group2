# Ejercicios sobre Pandas

```python
import pandas as pd
```

```python
# Crear DataFrame desde el archivo CSV
df_accidentes = pd.read_csv('Dataset/Nuclear_Incidents.csv')
df_accidentes.head(2)
```

```python
df_accidentes.columns
```

```python
# Eliminar espacios en blanco al principio y al final de los nombres de las columnas
df_accidentes.columns = df_accidentes.columns.str.strip()

# Cambiar el nombre de la columna 'INES\nlevel' a 'INES level'
df_accidentes.rename(columns={'INES\nlevel': 'INES level'}, inplace=True)

# Imprimir las primeras filas del DataFrame para verificar los cambios
df_accidentes
```

### Ejercicio 1:
* Muestra la cantidad de registros para cada país ('Location').

```python
# Extraer el país de la columna 'Location'
df_accidentes['Pais'] = df_accidentes['Location'].str.split(',').str[-1].str.strip()

# Mostrar la cantidad de incidentes por país
conteo_por_pais = df_accidentes['Pais'].value_counts()
conteo_por_pais
```

### Ejercicio 2:
* Eliminar las columnas 'Longitude', 'Latitude' y 'Location' y mostrar las 2 primeras filas del Dataframe.

```python
df_accidentes = df_accidentes.drop(['Longitude', 'Latitude', 'Location'], axis=1)
df_accidentes.head(2)
```

### Ejercicio 3:
* Muestra la cantidad de veces que se repiten los años en la columna 'Date'.

```python
# Mostrar la cantidad de registros en la columna 'Date' sin valores faltantes
conteo_fechas = pd.to_datetime(df_accidentes['Date']).dt.year.value_counts()
print(conteo_fechas)
```

### Ejercicio 4:
* Crea una nueva columna llamada 'Severity' que sea la suma de las columnas 'INES level' y 'Numbers of Direct Deaths'.

```python
# Crear la columna 'Severity'
df_accidentes['Severity'] = df_accidentes['INES level'] + df_accidentes['Numbers of Direct Deaths']

# Mostrar el DataFrame con la nueva columna
print(df_accidentes['Severity'])
```

### Ejercicio 5:
* Crear un nuevo campo, que se denomine "Total Deaths", que guarde la sumatoria de las columnas 'Numbers of Direct Deaths' y 'Numbers of InDirect Deaths'. Informar el total de muertes para el incidente más grave (mayor 'Total Deaths').

```python
# Reviso el contenido de la columna 'Numbers of Direct Deaths'
df_accidentes['Numbers of Direct Deaths'].info()
```

```python
df_accidentes['Numbers of InDirect Deaths'].info()
```

```python
df_accidentes['Numbers of InDirect Deaths'].value_counts()
```

```python
df_accidentes['Numbers of InDirect Deaths'] = df_accidentes['Numbers of InDirect Deaths'].astype(str)
```

```python
df_accidentes['Numbers of InDirect Deaths'] = df_accidentes['Numbers of InDirect Deaths'].replace('4,000', '4000')
```

```python
# Convertimos las columnas 'Numbers of Direct Deaths' y 'Numbers of InDirect Deaths' a tipo float
df_accidentes['Numbers of Direct Deaths'] = df_accidentes['Numbers of Direct Deaths'].astype(float)
df_accidentes['Numbers of InDirect Deaths'] = df_accidentes['Numbers of InDirect Deaths'].astype(float)
```

```python
df_accidentes['Total Deaths'] = df_accidentes['Numbers of Direct Deaths'] + df_accidentes['Numbers of InDirect Deaths']
max_total_deaths = df_accidentes.loc[df_accidentes['Total Deaths'].idxmax()]['Total Deaths']
```

### Ejercicio 8:
* Calcular la cantidad total de muertes directas por incidente nuclear para cada año.

```python
# Asegúrate de que la columna 'Date' sea del tipo datetime antes de continuar
df_accidentes['Date'] = pd.to_datetime(df_accidentes['Date'], format='%d/%m/%Y')

# Crea una nueva columna 'Year' que contenga el año de cada fecha
df_accidentes['Year'] = df_accidentes['Date'].dt.year

# Calculamos la cantidad de muertes por año
df_muertes_por_anio = df_accidentes.groupby('Year')['Numbers of Direct Deaths'].sum()

df_muertes_por_anio
```

### Ejercicio 9:
* Mostrar la cantidad de incidentes nucleares en cada país ('Pais') en los que el nivel INES es mayor a 5.

```python
conteo_altos_ines = df_accidentes[df_accidentes['INES level'] > 5]['Pais'].value_counts()
conteo_altos_ines
```

### Ejercicio 10:
* Calcular la cantidad de muertes directas totales de Estados Unidos.

```python
# Filtrar las filas para Estados Unidos y muertes directas mayores a 0
df_muertes_directas_usa = df_accidentes.loc[(df_accidentes['Pais'] == 'United States') & (df_accidentes['Numbers of Direct Deaths'] > 0)]

# Sumar la columna 'Numbers of Direct Deaths' para obtener el total
total_muertes_directas_usa = df_muertes_directas_usa['Numbers of Direct Deaths'].sum()

total_muertes_directas_usa
```
